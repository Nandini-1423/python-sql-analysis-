```python
import pandas as pd
df = pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
#rename column ..make them lower case and replace sapace with underscore
df.rename(columns={'Order Id':'order_id','City':'city'})
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.head(5)
#data.dtypes 
```

# derive new columns discount , sale price and profit 

```python
df.dtypes
```

```python
# 1. column of discount (list price * discount percent)
df['discount']=df['list_price']*df['discount_percent']*.01 

df.head()
```

```python
#2 . column of sale_price (list price - discount percent)
df['sale_price'] =df['list_price']-df['discount']
df
```

```python
# 3 . column of profit (sale price - cost_price)
df['profit'] =df['sale_price']-df['cost_price']
df
```

```python
  #convert order date from object data type to datetime
df.dtypes
```

```python
#you can see all the datatypes of date in the form of object , now you have to change it in date
pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
# you can drop cost price, list price and discount percent columns , bcz this is not needed 
df.drop(columns=['cost_price','list_price','discount_percent'] ,inplace=True)
```

```python
df.dtypes  # now you can see that these columns are permanenty delete 
```
